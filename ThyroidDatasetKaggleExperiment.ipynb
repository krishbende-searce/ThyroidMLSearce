```python
feature_indices = [0, 1, 2,7,9,10,11,12,13,14,15]
target_index = 16

X = catdata.iloc[:, feature_indices]
y = catdata.iloc[:, target_index]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Define the function to create a linear regression model using Keras
def build_model(learning_rate):
    model = Sequential()
    model.add(Dense(1, input_dim=X_train.shape[1], activation='linear'))
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=learning_rate), 
                  loss='mean_squared_error')
    return model

# Train the model
def train_model(model, X_train, y_train, epochs, batch_size):
    history = model.fit(X_train, y_train, epochs=epochs, batch_size=batch_size, verbose=1, validation_split=0.4)
    trained_weight = model.get_weights()[0]
    trained_bias = model.get_weights()[1]
    rmse = np.sqrt(history.history['loss'][-1])
    return trained_weight, trained_bias, history.epoch, rmse

# Set the hyperparameters
learning_rate = 0.07
epochs = 100
batch_size = 30

# Build and train the model
my_model = build_model(learning_rate)
trained_weight, trained_bias, epochs, rmse = train_model(my_model, X_train, y_train, epochs, batch_size)

# Evaluate the model
loss = my_model.evaluate(X_test, y_test, verbose=0)
print("Test Loss:", loss)

# Assuming predictions and y_test are defined
predictions = my_model.predict(X_test)
# Convert continuous predictions to binary class labels
threshold = 0.5
predicted_classes = (predictions > threshold).astype(int)
accuracy = accuracy_score(y_test, predicted_classes) * 100
print(f"Accuracy: {accuracy:.2f}%")
```

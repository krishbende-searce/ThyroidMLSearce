```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import random
```

```python
file_path = "C:/Users/krish/Desktop/ThyroidKaggleDataset/Thyroid_Diff.csv"
data = pd.read_csv(file_path)
```

```python
catenc=pd.get_dummies(data)
```

```python
catenc.info()
```

```python
X = catenc.iloc[:, 0:55]
y = catenc.iloc[:, 56]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

def build_model(learning_rate):
    model = Sequential()
    model.add(Dense(1, input_dim=X_train.shape[1], activation='linear'))
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=learning_rate), 
                  loss='mean_squared_error')
    return model

def train_model(model, X_train, y_train, epochs, batch_size):
    history = model.fit(X_train, y_train, epochs=epochs, batch_size=batch_size, verbose=1, validation_split=0.15)
    trained_weight = model.get_weights()[0]
    trained_bias = model.get_weights()[1]
    rmse = np.sqrt(history.history['loss'][-1])
    return trained_weight, trained_bias, history.epoch, rmse

learning_rate = 0.001
epochs = 300
batch_size = 60

my_model = build_model(learning_rate)
trained_weight, trained_bias, epochs, rmse = train_model(my_model, X_train, y_train, epochs, batch_size)

loss = my_model.evaluate(X_test, y_test, verbose=0)
print("Test Loss:", loss)
```
